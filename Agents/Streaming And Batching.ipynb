```python
import os
from dotenv import load_dotenv
load_dotenv(dotenv_path="../.env")
from IPython.display import display, Markdown

os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
os.environ["OPENROUTER_API_KEY"] = os.getenv("OPENROUTER_API_KEY")
os.environ["XAI_API_KEY"] = os.getenv("XAI_API_KEY")
```

```python
from langchain.chat_models import init_chat_model
model = init_chat_model(model="google_genai:gemini-3.1-flash-lite")
for chunk in model.stream("What SaaS business model can i start with genAI as a fresher in software developer"):
    if(len(chunk.content) > 0):
        response = chunk.content[0]["text"]
        display(Markdown(response))
```

```python
from langchain.chat_models import init_chat_model
model = init_chat_model(model="google_genai:gemini-3.1-flash-lite")
responses = model.batch([
    "Why do parratos talk",
    "What is the possible saas solution a fresher can make and turn it into a subsciption based SaaS"
])
for res in responses:
    display(Markdown(res.content[0]["text"]))
```
